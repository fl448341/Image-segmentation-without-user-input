```python
class SyntheticData[T](Dataset):
    """A small synthetic segmentation dataset.

    It is a sequence dataset of 5000 tuples (image, class, mask), where:
    - image: before transformation, an RGB PIL Image.
    - class: int 0..4, the label index.
    - mask:  numpy array of dtype=bool, shape (H, W), same size as image.
    """

    CLASSES: Final[tuple[str, ...]] = ("circle", "square", "triangle", "star", "diamond")

    def __init__(
        self,
        path: Path,
        split: Literal["train", "val", "test"],
        transform: Callable[[PIL.Image.Image], T],
    ) -> None:
        self.path = path
        self.transform = transform
        self.split = split

        self.class_to_idx = {class_name: i for i, class_name in enumerate(self.CLASSES)}

        images = sorted((self.path / "images" / self.split).glob("*.png"))
        masks = sorted((self.path / "masks" / self.split).glob("*.png"))
        labels = sorted((self.path / "labels" / self.split).glob("*.txt"))

        assert images, f"No images found in {self.path / 'images' / self.split}"
        assert len(images) == len(masks) == len(labels), (
            "Number of images, masks, and labels must be the same"
        )
        assert [p.stem for p in images] == [p.stem for p in masks] == [p.stem for p in labels], (
            "Image/mask/label filename mismatch."
        )
        self.image_names = [p.stem for p in images]

    def __len__(self) -> int:
        return len(self.image_names)

    def __getitem__(self, idx) -> tuple[T, int, np.ndarray]:
        name = self.image_names[idx]
        img_path = self.path / "images" / self.split / f"{name}.png"
        mask_path = self.path / "masks" / self.split / f"{name}.png"
        label_path = self.path / "labels" / self.split / f"{name}.txt"

        img = PIL.Image.open(img_path).convert("RGB")
        mask = np.array(PIL.Image.open(mask_path), dtype=bool)
        label = self.class_to_idx[label_path.read_text().strip()]

        img_transformed: T = self.transform(img)

        return img_transformed, label, mask


def show_image_row(
    image_dict: dict[str, PIL.Image.Image | np.ndarray | Tensor], size: float = 3.0
) -> None:
    n = len(image_dict)
    _, axs = plt.subplots(1, n, figsize=(size * n, size), constrained_layout=True, squeeze=True)
    if n == 1:
        axs = [axs]
    for ax, (title, img) in zip(axs, image_dict.items(), strict=True):
        ax.imshow(img)
        ax.set_title(title)
        ax.axis("off")
    plt.show()
```

```python
def train(
    model: nn.Module,
    train_loader: torch.utils.data.DataLoader,
    val_loader: torch.utils.data.DataLoader,
    test_loader: torch.utils.data.DataLoader,
    epochs: int = 5,
    **optimizer_kwargs: Any,
) -> None:
    print(f"ðŸš€ Training CNN for {epochs} epochs...")
    optimizer = optim.AdamW(model.parameters(), **optimizer_kwargs)

    for epoch in range(epochs):
        _train_epoch(model, train_loader, optimizer, desc=f"Epoch {epoch + 1}/{epochs} training  ")
        val_metrics = evaluate(model, val_loader, desc=f"Epoch {epoch + 1}/{epochs} validation")
        print(
            f"Epoch {epoch + 1}/{epochs} â€” "
            + f"val loss: {val_metrics['loss']:.3f}, val acc: {val_metrics['accuracy']:.1%}"
        )

    test_metrics = evaluate(model, test_loader, desc="Test Evaluation")
    print(
        "âœ… Model training complete: "
        + f"Test loss: {test_metrics['loss']:.3f}, test acc: {test_metrics['accuracy']:.1%}"
    )


def _train_epoch(
    model: nn.Module,
    dataloader: torch.utils.data.DataLoader,
    optimizer: optim.Optimizer,
    desc: str,
) -> dict[str, float]:
    model.train()
    device = next(model.parameters()).device
    total_loss = 0
    total_correct = 0
    total_samples = 0
    progress_bar = tqdm(dataloader, desc=desc)
    for imgs, labels, _ in progress_bar:
        imgs, labels = imgs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(imgs)
        loss = nn.CrossEntropyLoss()(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

        _, predicted = torch.max(outputs.data, dim=1)
        total_samples += labels.shape[0]
        total_correct += (predicted == labels).sum().item()

        progress_bar.set_postfix(
            train_loss=f"{total_loss / (total_samples / labels.shape[0]):.3f}",
            train_acc=f"{total_correct / total_samples:.1%}",
        )

    return {"loss": total_loss / len(dataloader), "accuracy": total_correct / total_samples}


def evaluate(
    model: nn.Module, dataloader: torch.utils.data.DataLoader, desc: str
) -> dict[str, float]:
    model.eval()
    device = next(model.parameters()).device
    total_loss = 0
    total_correct = 0
    total_samples = 0
    with torch.no_grad():
        for imgs, labels, _mask in tqdm(dataloader, desc=desc):
            imgs, labels = imgs.to(device), labels.to(device)
            outputs = model(imgs)
            loss = nn.CrossEntropyLoss()(outputs, labels)
            total_loss += loss.item()
            _, predicted = torch.max(outputs.data, dim=1)
            total_samples += labels.shape[0]
            total_correct += (predicted == labels).sum().item()

    return {"loss": total_loss / len(dataloader), "accuracy": total_correct / total_samples}


class DataloaderArgs(TypedDict, total=False):
    batch_size: int
    shuffle: bool
    num_workers: int
    pin_memory: bool
```

# 1. GradCAM

## Grad-CAM Implementation

In this section, we implement the `GradCAM` class. It uses PyTorch hooks (`register_forward_hook` and `register_full_backward_hook`) to capture feature maps and gradients from the target layers.

The implementation follows the logic:
1.  **Forward pass**: Capture activations.
2.  **Backward pass**: Capture gradients for the specific target class.
3.  **Weight computation**: Global Average Pooling of gradients.
4.  **Heatmap generation**: Weighted sum of activations followed by ReLU.
5.  **Normalization**: Min-max scaling to [0, 1] range, performed independently for each image.

```python
class GradCAM:
    """
    Pure PyTorch implementation of Grad-CAM.

    Usage:
        grad_cam = GradCAM(model=model, target_layers=[layer1, layer2])
        grayscale_cam = grad_cam(input_tensor, targets=[class_id])
    """
    def __init__(self, model: nn.Module, target_layers: Iterable[nn.Module]) -> None:
        self.model = model
        self.target_layers = target_layers
        self.activations = {}
        self.gradients = {}

        # Hooks for saving activations and gradients
        def save_activation(name):
            def hook(module, input, output):
                self.activations[name] = output
            return hook

        def save_gradient(name):
            def hook(module, grad_input, grad_output):
                self.gradients[name] = grad_output[0]
            return hook

        for i, layer in enumerate(self.target_layers):
            layer.register_forward_hook(save_activation(f'layer_{i}'))
            layer.register_full_backward_hook(save_gradient(f'layer_{i}'))

    def __call__(self, input_tensor: Tensor, targets: Iterable[int] | None = None) -> np.ndarray:
        self.model.eval()
        self.model.zero_grad()

        # Forward pass
        output = self.model(input_tensor)

        if targets is None:
            targets = output.argmax(dim=1)

        # Backward pass
        one_hot = torch.zeros_like(output)
        for i, target in enumerate(targets):
            one_hot[i, target] = 1.0
        output.backward(gradient=one_hot, retain_graph=True)

        heatmaps_normalized_list = []

        # 3. Compute CAM for each layer
        for i in range(len(self.target_layers)):
            name = f'layer_{i}'
            grads = self.gradients[name] # [B, C, H, W]
            acts = self.activations[name] # [B, C, H, W]

            # Weights (Global Average Pooling on gradients)
            weights = torch.mean(grads, dim=(2, 3), keepdim=True)

            cam = torch.sum(weights * acts, dim=1, keepdim=True)
            cam = F.relu(cam)

            cam = F.interpolate(cam, size=(input_tensor.shape[2], input_tensor.shape[3]),
                                mode='bilinear', align_corners=False)
            cam = cam.squeeze(1) # [B, H, W]

            # Normalize "independently for each image and each target_layers"
            # Processing batch-wise for safety and readability
            cam_np = cam.detach().cpu().numpy()
            norm_cam_batch = []
            for img_cam in cam_np:
                if img_cam.max() - img_cam.min() > 0:
                    img_cam = (img_cam - img_cam.min()) / (img_cam.max() - img_cam.min())
                else:
                    img_cam = np.zeros_like(img_cam)
                norm_cam_batch.append(img_cam)

            heatmaps_normalized_list.append(np.array(norm_cam_batch))

        # Averaging (if more than one layer)
        if len(heatmaps_normalized_list) > 1:
            # Stack [L, B, H, W] and take the mean across axis 0
            final_cam = np.mean(np.stack(heatmaps_normalized_list), axis=0)
        else:
            final_cam = heatmaps_normalized_list[0]

        return final_cam
```

```python
def heatmap_to_rgb_image(
    heatmap: np.ndarray, min: float | None = None, max: float | None = None
) -> PIL.Image.Image:
    """
    Converts a single-channel heatmap to an RGB pillow image using a colormap.

    Args:
    - heatmap: shape (H, W), will be normalized by mapping min..max to 0..1.
    - min: minimum value for normalization, defaults to heatmap.min().
    - max: maximum value for normalization, defaults to heatmap.max()
    """
    heatmap = heatmap.astype(np.float32)
    if min is None:
        min = heatmap.min()
    if max is None:
        max = heatmap.max()
    heatmap = (heatmap - min) / (max - min + 1e-8)
    heatmap_uint8 = (np.clip(heatmap, 0.0, 1.0) * 255).astype(np.uint8)
    heatmap_color = cv2.applyColorMap(heatmap_uint8, cv2.COLORMAP_JET)
    heatmap_color = cv2.cvtColor(heatmap_color, cv2.COLOR_BGR2RGB)
    return PIL.Image.fromarray(heatmap_color)
```

```python
class BasicSamPipeline:
    def __call__(self, images: Tensor) -> Tensor:
        """
        Input: normalized images, shape (B, C=3, H, W).
        Output: masks tensor of shape (B, H, W), dtype=bool.
        """
        B, C, H, W = images.shape
        # The basic pipeline always uses a single center point for each image.
        point_coords = np.array([[(W // 2, H // 2)] for _ in range(B)])

        # The basic pipeline always uses a single foreground point, no background points.
        point_labels = np.array([[1] for _ in range(B)], dtype=np.int64)

        return self.segment(images, point_coords, point_labels)

    def segment(
        self, images: Tensor, point_coords: np.ndarray, point_labels: np.ndarray
    ) -> Tensor:
        """
        Args:
        - images: normalized images, shape (B, C=3, H, W).
        - point_coords: point coordinates within each image, shape (B, num_points, 2), format (x,y).
            Note the format is not (h,w)=(y,x), but (x,y)!
        - point_labels: point labels, shape (B, num_points), dtype int64.
            Label 1 is foreground (should be in mask), 0 is background (shouldn't be in mask).

        Returns: segmentation masks, shape (B, H, W), dtype=bool.
        """
        B, C, H, W = images.shape
        assert C == 3, f"Expected images.shape=(B, C=3, H, W), got: {images.shape}"
        num_points = point_coords.shape[1]
        assert point_coords.shape == (B, num_points, 2), f"Expected point_coords.shape=({B=}, num_points, 2), got: {point_coords.shape}"
        assert point_labels.shape == (B, num_points), f"Expected point_labels.shape=({B=}, num_points), got: {point_labels.shape}"

        results = list[Tensor]()
        for image, pt_coords, pt_labels in zip(images, point_coords, point_labels, strict=True):
            sam_predictor.set_image(np.array(inverse_transform(image)))
            masks, scores, _logits = sam_predictor.predict(
                point_coords=pt_coords, point_labels=pt_labels, multimask_output=True
            )
            best_mask = masks[np.argmax(scores)]
            results.append(torch.tensor(best_mask, dtype=torch.bool))
        return torch.stack(results)
```

## SAM Pipelines Implementation

Here we define two subclasses of `BasicSamPipeline` to automatically select prompts for the Segment Anything Model (SAM) based on Grad-CAM heatmaps.

1.  **`ForegroundSamPipeline`**: Selects a single foreground point corresponding to the **maximum** activation in the Grad-CAM heatmap.
2.  **`ForeBackgroundSamPipeline`**: Selects two points:
    * One foreground point (at the **maximum** activation).
    * One background point (at the **minimum** activation), helping SAM distinguish the object from the background.

```python
class ForegroundSamPipeline(BasicSamPipeline):
    """
    Approach 1: Foreground point only.
    We take the point with maximum activation from Grad-CAM.
    """
    def __init__(self, model_classifier, target_layers):
        self.grad_cam = GradCAM(model_classifier, target_layers)

    def __call__(self, images: Tensor) -> Tensor:
        B, C, H, W = images.shape

        # Generate Grad-CAM heatmaps for model predictions (without targets = top1 class)
        heatmaps = self.grad_cam(images, targets=None) # shape (B, H, W)

        point_coords = []
        point_labels = []

        for heatmap in heatmaps:
            flat_idx = np.argmax(heatmap)
            y, x = np.unravel_index(flat_idx, heatmap.shape)

            # SAM requires format (x, y)
            point_coords.append([(x, y)])
            point_labels.append([1])

        point_coords = np.array(point_coords)
        point_labels = np.array(point_labels, dtype=np.int64)

        return self.segment(images, point_coords, point_labels)


class ForeBackgroundSamPipeline(BasicSamPipeline):
    """
    Approach 2: Foreground point AND Background point.
    Foreground: Max Grad-CAM
    Background: Min Grad-CAM
    """
    def __init__(self, model_classifier, target_layers):
        self.grad_cam = GradCAM(model_classifier, target_layers)

    def __call__(self, images: Tensor) -> Tensor:
        B, C, H, W = images.shape

        heatmaps = self.grad_cam(images, targets=None)

        point_coords = []
        point_labels = []

        for heatmap in heatmaps:
            # Foreground Point (Max)
            flat_idx_max = np.argmax(heatmap)
            y_max, x_max = np.unravel_index(flat_idx_max, heatmap.shape)

            # Background Point (Min) - where the model attends the least
            flat_idx_min = np.argmin(heatmap)
            y_min, x_min = np.unravel_index(flat_idx_min, heatmap.shape)

            point_coords.append([(x_max, y_max), (x_min, y_min)])
            # Label 1: object, Label 0: background
            point_labels.append([1, 0])

        point_coords = np.array(point_coords)
        point_labels = np.array(point_labels, dtype=np.int64)

        return self.segment(images, point_coords, point_labels)
```

## Evaluation Metrics

This section contains helper functions to evaluate the performance of the pipelines:
* **`compute_iou`**: Calculates Intersection over Union between the predicted mask and the ground-truth mask.
* **`evaluate_pipeline`**: Iterates over the test dataset, computes metrics, and reports:
    * **Mean IoU**: Overall segmentation quality (Target: >65%).
    * **Hit Rate**: Percentage of generated foreground points falling inside the ground-truth mask.
    * **Mean Distance**: Average distance (in pixels) of the foreground point from the center of mass of the ground-truth object.

```python
def compute_iou(pred_mask, gt_mask):
    intersection = torch.logical_and(pred_mask, gt_mask).sum().item()
    union = torch.logical_or(pred_mask, gt_mask).sum().item()
    if union == 0:
        return 0.0 if intersection == 0 else 1.0
    return intersection / union

def evaluate_pipeline(pipeline, dataloader, name="Pipeline"):
    print(f"--- Evaluating {name} ---")
    ious = []
    hit_rates = [] # For foreground point
    distances = [] # For foreground point

    # Access internal Grad-CAM to check where the point landed.
    # In a normal pipeline, __call__ doesn't return points, so we must reconstruct logic
    # or modify pipeline classes, but we keep the __call__ signature intact.
    # We use a 'brute force' method - recreating point selection logic inside the loop.

    grad_cam_engine = pipeline.grad_cam

    for images, labels, gt_masks in tqdm(dataloader):
        images = images.to(device)
        gt_masks = gt_masks.to(device) # shape (B, H, W)

        # Mask prediction
        pred_masks = pipeline(images).to(device) # shape (B, H, W)

        for pred, gt in zip(pred_masks, gt_masks):
            ious.append(compute_iou(pred, gt))

        # Re-calculate heatmap to know which point was selected
        heatmaps = grad_cam_engine(images, targets=None)

        # Convert gt_masks to CPU numpy for easy geometric calculations
        gt_masks_np = gt_masks.cpu().numpy()

        for i, heatmap in enumerate(heatmaps):
            # Foreground Point (same as in pipeline logic)
            flat_idx = np.argmax(heatmap)
            y, x = np.unravel_index(flat_idx, heatmap.shape) # y=row, x=col

            gt_mask_i = gt_masks_np[i]

            is_hit = gt_mask_i[y, x]
            hit_rates.append(1.0 if is_hit else 0.0)

            # Distance from center of mass
            true_y, true_x = np.where(gt_mask_i)
            if len(true_y) > 0:
                center_y = np.mean(true_y)
                center_x = np.mean(true_x)
                dist = np.sqrt((y - center_y)**2 + (x - center_x)**2)
                distances.append(dist)
            else:
                pass

    print(f"Results for {name}:")
    print(f"  Mean IoU: {np.mean(ious):.2%}")
    print(f"  Hit Rate: {np.mean(hit_rates):.2%}")
    print(f"  Mean Distance from Center: {np.mean(distances):.2f} px")
    print("\n")
```
